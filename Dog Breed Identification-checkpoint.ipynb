```python
%matplotlib inline
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
from os import listdir, makedirs
from os.path import join, exists, expanduser
from tqdm import tqdm

import keras
from sklearn.metrics import log_loss, accuracy_score
from keras.preprocessing import image
from keras.applications.vgg16 import VGG16
from keras.applications.resnet50 import ResNet50
from keras.applications import xception
from keras.applications import inception_v3
from keras.applications.vgg16 import preprocess_input, decode_predictions
from sklearn.linear_model import LogisticRegression
```

```python
INPUT_SIZE = 200
NUM_CLASSES = 120
SEED = 1987
data_dir = '/home/lab276/Downloads/dog/'
labels = pd.read_csv(join(data_dir, 'labels.csv'))
sample_submission = pd.read_csv(join(data_dir, 'sample_submission.csv'))
```

```python
selected_breed_list = list(labels.groupby('breed').count().sort_values(by='id', ascending=False).head(NUM_CLASSES).index)
labels = labels[labels['breed'].isin(selected_breed_list)]

labels['target'] = 1

labels['rank'] = labels.groupby('breed').rank()['id']

labels_pivot = labels.pivot('id', 'breed', 'target').reset_index().fillna(0)
#print(labels['target'])

np.random.seed(seed=SEED)
#print(labels_pivot)
rnd = np.random.random(len(labels))
train_idx = rnd < 0.8
valid_idx = rnd >= 0.8
#print(valid_idx)
y_train = labels_pivot[selected_breed_list].values
Y_train = y_train[train_idx]
Y_val = y_train[valid_idx]
#print(ytr)
#print(yv)
```

```python
def read_img(img_id, train_or_test, size):
    
    img = image.load_img(join(data_dir, train_or_test, '%s.jpg' % img_id), target_size=size)
    img = image.img_to_array(img)
    return img
```

```python
datagen = keras.preprocessing.image.ImageDataGenerator(
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True)
```

```python
print(len(labels))
```

```python

INPUT_SIZE = 299
POOLING = 'avg'
X_train = np.zeros((len(labels), INPUT_SIZE, INPUT_SIZE, 3), dtype='float32')
for i, img_id in tqdm(enumerate(labels['id'])):
    img = read_img(img_id, 'train', (INPUT_SIZE, INPUT_SIZE))
    x = xception.preprocess_input(np.expand_dims(img.copy(), axis=0))
    X_train[i] = x
    
datagen.fit(X_train)
print('Train Images shape: {} size: {:,}'.format(X_train.shape, X_train.size))
```

```python
X_tr = X_train[train_idx]
X_val = X_train[valid_idx]
print((X_tr.shape, X_val.shape, Y_train.shape, Y_val.shape))
xception_bottleneck = xception.Xception(weights='imagenet', include_top=False, pooling=POOLING)
inception_features = keras.applications.inception_resnet_v2.InceptionResNetV2(include_top=False, weights='imagenet',pooling=POOLING)
vgg19_features = keras.applications.vgg19.VGG19(include_top=False, weights='imagenet',pooling=POOLING)


train_x_bf_x = xception_bottleneck.predict(X_tr, batch_size=32, verbose=1)
valid_x_bf_x = xception_bottleneck.predict(X_val, batch_size=32, verbose=1)

train_x_bf_in = inception_features.predict(X_tr, batch_size=32, verbose=1)
valid_x_bf_in = inception_features.predict(X_val, batch_size=32, verbose=1)

#train_x_bf_v = vgg19_features.predict(Xtr, batch_size=32, verbose=1)
#valid_x_bf_v = vgg19_features.predict(Xv, batch_size=32, verbose=1)

#print('Xception train bottleneck features shape: {} size: {:,}'.format(train_x_bf.shape, train_x_bf.size))
#print('Xception valid bottleneck features shape: {} size: {:,}'.format(valid_x_bf.shape, valid_x_bf.size))
```

```python
from keras.layers import *
from keras.models import *
from keras.applications import *
from keras.optimizers import *
from keras.regularizers import *


train_x_bf = np.concatenate([train_x_bf_x, train_x_bf_in], axis=-1)
valid_x_bf = np.concatenate([valid_x_bf_x, valid_x_bf_in], axis=-1)

inputs = Input(train_x_bf.shape[1:])
x = inputs
x = Dropout(0.95)(x)
x = Dense(120, activation='softmax')(x)
model = Model(inputs, x)
model.compile(optimizer='sgd',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
sgd = optimizers.SGD(lr=0.0001, decay=1e-10)#, momentum=0.95, nesterov=True)

h = model.fit(train_x_bf, Y_train, batch_size=256, epochs=600, validation_split=0.1)
```
